```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('../vehicles_us.csv') # carga de datos
fig = px.histogram(data, x="odometer") # crear un histograma
fig.show() # crear gráfico de dispersión´
```

```python
fig = px.scatter(data, x="odometer", y="price") # crear un gráfico de dispersión
fig.show() # crear gráfico de dispersión
```

```python
filtered_data = data.dropna(subset=["cylinders"])
filtered_data = filtered_data[filtered_data["price"] > 0]
fig = px.scatter(filtered_data, x="model_year", y="price",
	         size="cylinders", color="fuel",
                 hover_name="type", log_x=True, size_max=60)
fig.show()
```
